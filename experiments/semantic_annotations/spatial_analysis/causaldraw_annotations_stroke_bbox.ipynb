```python
import os
import pandas as pd
import numpy as np
import cv2
import glob
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
base_dir = './../'
csv_dir = os.path.join(base_dir, "results", "csv")
roi_dir = os.path.join(base_dir, "results", "roi")
df = pd.read_csv(os.path.join(csv_dir, "causaldraw_annotation_preprocessed_final_svg_data.csv"), index_col=0)
```

```python
df = df[["sketchID", "strokeRoiNum", "condition", "orig_gameID", "strokeLabel", "strokeType"]]
df = df.drop_duplicates()
df = df.fillna(0)
df
```

```python
for i, row in tqdm(df.iterrows(), total=len(df)):
    
    sketch_id = row["sketchID"].replace("_", "-")
    stroke_roi = int(row["strokeRoiNum"])
    
    filename = os.path.join(roi_dir, "%s_%d.png" % (sketch_id, stroke_roi))
    img = cv2.imread(filename, cv2.IMREAD_UNCHANGED)
    roi = np.where(img != 0)
    x1, x2, y1, y2 = np.min(roi[1]), np.max(roi[1]), np.min(roi[0]), np.max(roi[0])
    df.loc[i, "draw_x1"] = x1
    df.loc[i, "draw_x2"] = x2
    df.loc[i, "draw_y1"] = y1
    df.loc[i, "draw_y2"] = y2
```

```python
df.to_csv(os.path.join(csv_dir, "causaldraw_annotation_preprocessed_final_bbox.csv"))
```
